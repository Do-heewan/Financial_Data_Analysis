### 01 데이터 로드

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drink_df = pd.read_csv(url)
drink_df
```

### 02 데이터 로직

```python
# 데이터프레임의 이름.특정 열의 이름 == '특정값' -> boolean
drink_df.continent == 'EU'
```

```python
drink_df['continent'] == 'EU'
```

```python
type(drink_df.continent == 'EU')
```

```python
# 데이터프레임[true, false로 구성된 시리즈] -> true만 반환
drink_df[drink_df.continent == 'EU'].head()
```

```python
drink_df.beer_servings > 158
```

```python
drink_df[drink_df.beer_servings > 158].head()
```

```python
# 아래 두 식의 결과는 같다
print(drink_df[drink_df.beer_servings > 158].head(10))
print(drink_df[drink_df['beer_servings'] > 158].head(10))
```

```python
# 데이터프레임의 이름[['특정 열의 이름1, '특정 열의 이름2']]
# beer_servings의 값이 10 이하이면서 country, beer_servings의 두 개의 열만을 뽑아내는 방법

drink_df[drink_df.beer_servings <= 10][['country', 'beer_servings']].head()
```

```python
# 조건으로 필터링하여 시리즈(Series)를 얻어낸 뒤에 평균(mean)을 얻어보기
drink_df[drink_df.continent=='EU'].beer_servings.mean()
```

```python
drink_df[drink_df.beer_servings > drink_df.beer_servings.mean()] # drink_df.beer_servings.mean() = 106
```

```python
# isnull()은 해당 행에 Null 값이 들어있는 경우만 True를 리턴하고, 아니라면 False를 리턴
drink_df.continent.isnull()
```

```python
drink_df.continent.isnull().sum()
```

```python
drink_df[drink_df.continent.isnull()]
```

```python
len(drink_df[drink_df.continent.isnull()])
```
