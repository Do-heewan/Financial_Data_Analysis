### 01 datetime64

```python
import numpy as np
np.datetime64('2024-10-30') # numpy객체를 이용해 날짜를 생성
```

```python
import numpy as np

# 나노초 단위
np.datetime64(1000, 'ns')
```

```python
np.datetime64(10000000000, 'ns')
```

```python
np.datetime64(10000000000, 's')
```

```python
np.datetime64(1000, 'D')
```

### 02 array()

```python
np.array(['2007-07-13', '2006-01-13', '2010-08-13'], dtype = 'datetime64')
```

```python
# 달단위
np.array(['2007-07-13', '2006-01-13', '2010-08-13'], dtype = 'datetime64[M]')
```

### 03 arange()

```python
# 달단위
np.arange('2005-02', '2006-03', dtype = 'datetime64[M]')
```

```python
# 주단위
np.arange('2005-02', '2006-03', dtype = 'datetime64[W]')
```

```python
# 일단위
np.arange('2005-02', '2006-03', dtype = 'datetime64[D]')
```

```python
# 시간단위
np.arange('2005-02', '2006-03', dtype = 'datetime64[h]')
```

### 04 datetime64를 이용한 날짜 차이 구하기

```python
# 2009년 1월 1일 - 2008년 1월
np.datetime64('2009-01-01') - np.datetime64('2008-01')
```

```python
# 2009년 1월 1일 - 2008년 1월 1일
np.datetime64('2009-01-01') - np.datetime64('2008-01-01')
```

```python
# 2009년 - 2008년 1월
np.datetime64('2009') - np.datetime64('2008-01')
```
